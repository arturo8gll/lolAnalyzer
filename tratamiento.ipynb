```python
def test_chisquare(df1,df2):
    table = pd.DataFrame()
    pvalue = []
    si = []
    variables=separateVarCD(df1)
    for i in variables[1]:
        pvalue.append(chisquare(df1[i].value_counts().sort_index(),df2[i].value_counts().sort_index()))
        si.append('true' if chisquare(df1[i].value_counts().sort_index(),df2[i].value_counts().sort_index())[1] <.05 else'false')
    table['variable'] = variables[0]
    table['p_value']=pvalue
    table['son iguales']=si
    return table
```
